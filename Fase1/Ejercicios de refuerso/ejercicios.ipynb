# Ejercicio 1 — Lectura consciente (SELECT con sentido)
Objetivo: Confirmar que entiendes qué información estás pidiendo y por qué.

Qué debes lograr

____Obtener información de una sola tabla

____No traer “todo”, sino solo lo que tiene sentido

____Limitar resultados para inspección

Conceptos que estás practicando

____selección de columnas

____lectura parcial

____control de volumen de datos

Pregunta clave (para ti)

¿Por qué traerías solo 5 registros y no 1000 cuando estás explorando?
